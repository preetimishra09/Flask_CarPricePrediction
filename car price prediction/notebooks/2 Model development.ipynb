```python
import pandas as pd
```

```python
df = pd.read_csv('../data/processed data/cleaned_data_v1.csv')
df.head()
```

```python
X=df.drop(['new_Selling_Price'],axis=1)
X
```

```python
y=df['new_Selling_Price']
y
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=12)
```

```python
type(X_test)
```

```python
X_test.to_csv("../data/test data/test_data.csv",index=False)
```

```python
# Fitting Simple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
import pickle
with open('../models/lr_model.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)
```

```python
ypred=model.predict(X_test)
```
